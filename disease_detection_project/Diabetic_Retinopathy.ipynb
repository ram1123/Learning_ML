## Step 1.4 : Calculate quantile and plot Quantile-Quantile (Q-Q) plot
*This gives us idea about the outliers*

```python
# Ref: https://docs.scipy.org/doc/scipy-0.19.1/reference/generated/scipy.stats.probplot.html
# Ref: http://www.statisticshowto.com/q-q-plots/#comment-609679

import numpy as np 
import pylab 
import scipy.stats as stats
import matplotlib.pyplot as plt

for columns in infile:
    fig = plt.figure()
    pylab = fig.add_subplot(111)
    stats.probplot(infile[columns], dist="norm", plot=pylab)
    pylab.set_title(columns)
    plt.show()
```
