# Standardization and Min-Max Scaling
Reference: http://nbviewer.jupyter.org/github/rasbt/pattern_classification/blob/master/preprocessing/about_standardization_normalization.ipynb

```python
from sklearn import preprocessing

std_scale = preprocessing.StandardScaler().fit(infile[['qua_ass','pre_screening','MA_CL05','MA_CL06','MA_CL07','MA_CL08','MA_CL09','MA_CL10','MA_exu_CL05',
                  'MA_exu_CL06','MA_exu_CL07','MA_exu_CL08','MA_exu_CL09','MA_exu_CL10','MA_exu_CL11','MA_exu_CL12',
                  'distance','diameter','AM/FM']])
df_std = std_scale.transform(infile[['qua_ass','pre_screening','MA_CL05','MA_CL06','MA_CL07','MA_CL08','MA_CL09','MA_CL10','MA_exu_CL05',
                  'MA_exu_CL06','MA_exu_CL07','MA_exu_CL08','MA_exu_CL09','MA_exu_CL10','MA_exu_CL11','MA_exu_CL12',
                  'distance','diameter','AM/FM']])

minmax_scale = preprocessing.MinMaxScaler().fit(infile[['qua_ass','pre_screening','MA_CL05','MA_CL06','MA_CL07','MA_CL08','MA_CL09','MA_CL10','MA_exu_CL05',
                  'MA_exu_CL06','MA_exu_CL07','MA_exu_CL08','MA_exu_CL09','MA_exu_CL10','MA_exu_CL11','MA_exu_CL12',
                  'distance','diameter','AM/FM']])
df_minmax = minmax_scale.transform(infile[['qua_ass','pre_screening','MA_CL05','MA_CL06','MA_CL07','MA_CL08','MA_CL09','MA_CL10','MA_exu_CL05',
                  'MA_exu_CL06','MA_exu_CL07','MA_exu_CL08','MA_exu_CL09','MA_exu_CL10','MA_exu_CL11','MA_exu_CL12',
                  'distance','diameter','AM/FM']])
```

```python
print('Mean after standardization:\nqua_ass ={:.2f}, pre_screening={:.2f}'
      .format(df_std[:,0].mean(), df_std[:,1].mean()))
print('\nStandard deviation after standardization:\nqua_ass ={:.2f}, pre_screening={:.2f}'
      .format(df_std[:,0].std(), df_std[:,1].std()))
```

# STEP: 2 : Apply some machine learning algorithm (MLA)

# Data Slicing and Train-Test split
Below we defined two variables x and y. Here we are splitting our data into features and labels. Y contains labels (i.e. it is result that a patient has diabetic retinopathy or not), and X contains list of all features.

```python
X = infile.values[:,0:19]
Y = infile.values[:,19]
```

### Below we are going to split the X and Y into training and test set. So, using X we will train our algorithm and after training we will test its aplicability using Y.

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state=100)
```

### Check shape of training and test dataset.

```python
print "X_train shape = ",X_train.shape
print "X_test shape = ",X_test.shape
print "Y_train shape = ",y_train.shape
print "Y_test shape = ",y_test.shape
```

## Apply BDT 
Ref: http://dataaspirant.com/2017/02/01/decision-tree-algorithm-python-with-scikit-learn/

```python
#Import modules for decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

## BDT optimization: Iteration 1

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train,y_train)
```

```python
clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=100,
                                    max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred = clf_gini.predict(X_test)
y_pred
```

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

```python
print "Accuracy is ",accuracy_score(y_test, y_pred)*100
```

```python
print "Accuracy with information gain is ",accuracy_score(y_test, y_pred_en)*100
```

## BDT optimization: Scan over number of leaf and branches

```python
tmp_score = 0.0
ndepth = 0
nleaf = 0
for depth in range(1,20):
    for leaf in range(1,20):
        clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=depth, min_samples_leaf=leaf)
        clf_gini.fit(X_train,y_train)
        y_pred = clf_gini.predict(X_test)
        print "Depth = ",depth,"\t Leaf = ",leaf,"\t Accuracy is ",accuracy_score(y_test, y_pred)*100
        if tmp_score < (accuracy_score(y_test, y_pred)*100):
            tmp_score = (accuracy_score(y_test, y_pred)*100)
            ndepth = depth
            nleaf = leaf
    print "\n==============================\n"
print "Final Result of scan:"
print "===> Depth = ",ndepth,"\t leaf = ",nleaf,"\t accuracy score = ",tmp_score
```

```python
tmp_score = 0.0
ndepth = 0
nleaf = 0
for depth in range(1,20):
    for leaf in range(1,20):
        clf_gini = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth=depth, min_samples_leaf=leaf)
        clf_gini.fit(X_train,y_train)
        y_pred = clf_gini.predict(X_test)
        print "Depth = ",depth,"\t Leaf = ",leaf,"\t Accuracy is ",accuracy_score(y_test, y_pred)*100
        if tmp_score < (accuracy_score(y_test, y_pred)*100):
            tmp_score = (accuracy_score(y_test, y_pred)*100)
            ndepth = depth
            nleaf = leaf
    print "\n==============================\n"
print "Final Result of scan:"
print "===> Depth = ",ndepth,"\t leaf = ",nleaf,"\t accuracy score = ",tmp_score
```
